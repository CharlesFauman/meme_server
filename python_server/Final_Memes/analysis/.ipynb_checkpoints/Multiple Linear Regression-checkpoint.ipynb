```python
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 5
d = 3
max_value = 10
points = np.random.random((n, d))*max_value
X = np.column_stack([points[:,:-1], np.ones(n)])
Y = points[:,-1]
w = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, Y))
```

```python
print("X:", X)
print("w.T:", w.T)
print("Y':", np.dot(X, w.T))
print("Y:", Y)
print("E:", np.subtract(np.dot(X, w.T), Y))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for point in points:
    ax.scatter(X[:,0], X[:,1], Y, c='r', marker='o')

def fun(x1, x2):
    return w[0] * x1 + w[1] * x2 + w[2]

x1 = x2 = np.linspace(0, max_value)
X1, X2 = np.meshgrid(x1, x2)
y_ = np.array([fun(x1, x2) for x1, x2 in zip(np.ravel(X1), np.ravel(X2))])
Y_ = y_.reshape(X1.shape)

ax.plot_surface(X1, X2, Y_)

    
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.show()
```

```python
plt.figure()
plt.scatter(X[:,0], Y, c = 'r')
abline_values = [w[0] * i + w[2] for i in np.linspace(0, max_value)]
plt.plot(np.linspace(0, max_value), abline_values, 'b')
plt.xlabel('X1')
plt.ylabel('Y')
plt.show()

plt.figure()
plt.scatter(X[:,1], Y, c = 'r')
abline_values = [w[1] * i + w[2] for i in np.linspace(0, max_value)]
plt.plot(np.linspace(0, max_value), abline_values, 'b')
plt.xlabel('X2')
plt.ylabel('Y')
plt.show()
```
